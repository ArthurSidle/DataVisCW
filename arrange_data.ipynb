```python
import numpy as np
import pandas as pd
```

```python
companies = pd.read_csv("data/igdb_companies.csv")
involved_companies = pd.read_csv("data/igdb_involved_companies.csv")
games = pd.read_csv("data/igdb.csv")
external_games = pd.read_csv("data/igdb_external_games.csv")
developed = pd.read_csv("data/igdb_developed.csv")
steam_lost = pd.read_csv("data/steam_lost.csv", encoding="unicode_escape")
```

```python
external_games = external_games.query("platform == 'steam'")
com_df = games.merge(external_games[["game_id", "uid"]], on="game_id", how="inner")
com_df
```

```python
com_df["uid"] = pd.to_numeric(com_df["uid"])
com_df = com_df.merge(steam_lost, left_on="uid", right_on="AppID", how="inner")
com_df = com_df[["game_id","name","rating","rating_count","category","uid"]]
com_df
```

```python
com_df = com_df.merge(developed, on="game_id", how="inner").merge(companies, on="company_id", how="inner")
com_df.rename(columns={"name_x": "game_name", "name_y": "company_name"}, inplace=True)
com_df
```

```python
com_df["company_name"].value_counts().to_csv("final_data/serial_delisters.csv")
```

```python
genres = pd.read_csv("data/igdb_genres.csv")

gen_df = games.merge(external_games[["game_id", "uid"]], on="game_id", how="inner")
gen_df["uid"] = pd.to_numeric(gen_df["uid"])
gen_df = gen_df.merge(steam_lost, left_on="uid", right_on="AppID", how="inner")
#genres_simple = genres.drop_duplicates("game_id")
gen_df = gen_df.merge(genres, on="game_id", how="inner")
gen_df["genre"] = gen_df["genre"].replace(np.NaN, "Unknown")
gen_df
```

```python
gen_df["genre"].value_counts()
```

```python
gen_df["genre"].value_counts().to_csv("final_data/genres.csv")
```
