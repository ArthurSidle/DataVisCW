```python
import numpy as np
import pandas as pd
```

```python
companies = pd.read_csv("data/igdb_companies.csv")
involved_companies = pd.read_csv("data/igdb_involved_companies.csv")
games = pd.read_csv("data/igdb.csv")
external_games = pd.read_csv("data/igdb_external_games.csv")
developed = pd.read_csv("data/igdb_developed.csv")
steam_lost = pd.read_csv("data/steam_lost.csv", encoding="unicode_escape")
```

```python
external_games = external_games.query("platform == 'steam'")
df = games.merge(external_games[["game_id", "uid"]], on="game_id", how="inner")
df
```

```python
df["uid"] = pd.to_numeric(df["uid"])
df = df.merge(steam_lost, left_on="uid", right_on="AppID", how="inner")
df = df[["game_id","name","rating","rating_count","category","uid"]]
df
```

```python
df = df.merge(developed, on="game_id", how="inner").merge(companies, on="company_id", how="inner")
df.rename(columns={"name_x": "game_name", "name_y": "company_name"}, inplace=True)
df
```

```python
df["company_name"].value_counts().to_csv("final_data/serial_delisters.csv")
```
